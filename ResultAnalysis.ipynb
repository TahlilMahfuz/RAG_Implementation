```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('1.distilbert.csv')
df2 = pd.read_csv('2.roberta.csv')
df3 = pd.read_csv('3.robertaXL.csv')
df4 = pd.read_csv('4.McDSentimentAnalysis.csv')
df5 = pd.read_csv('5.multiLingual.csv')
df6 = pd.read_csv('6.AspectBasedSentimentAnalysis.csv')
df7 = pd.read_csv('6.AspectBasedSentimentAnalysis_withApplicability.csv')
```

```python
# List of DataFrames and their names
dataframes = [
    (df1, "DistilBERT"),
    (df2, "RoBERTa"),
    (df3, "RoBERTa XL"),
    (df4, "McD Sentiment Analysis"),
    (df5, "Multilingual"),
    (df6, "Aspect-Based Sentiment Analysis"),
    (df7, "Aspect-Based Sentiment Analysis with Applicability"),
]
```

```python
# Function to analyze and plot sentiment distribution with percentages
def analyze_and_plot_sentiment(df, title):
    # Check if the DataFrame is not empty
    if df.empty:
        print(f"{title} DataFrame is empty. Skipping...")
        return
    
    # Check if the 'Sentiment' column exists
    if 'Sentiment' not in df.columns:
        print(f"'Sentiment' column not found in {title}. Skipping...")
        return
    
    # Display sentiment distribution
    sentiment_counts = df['Sentiment'].value_counts()
    total = sentiment_counts.sum()
    percentages = (sentiment_counts / total) * 100
    print(f"Sentiment distribution for {title}:")
    print(sentiment_counts)
    print(f"Percentages:\n{percentages}")
    
    # Plot sentiment distribution
    plt.figure(figsize=(8, 6))
    ax = sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="viridis")
    plt.title(f"{title} - Sentiment Distribution")
    plt.xlabel("Sentiment")
    plt.ylabel("Count")
    
    # Annotate bars with percentages
    for i, count in enumerate(sentiment_counts.values):
        percentage = f"{percentages.iloc[i]:.1f}%"
        ax.text(i, count + 0.5, percentage, ha='center', va='bottom', fontsize=10, color='black')
    
    plt.show()

# Generate sentiment distribution diagrams for each DataFrame
for df, name in dataframes:
    analyze_and_plot_sentiment(df, name)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df7 is already loaded into your environment

# Set up the plotting environment
sns.set(style="whitegrid")

# Aspect Categories
aspects = ['feature', 'ads', 'performance']
labels = ['positive', 'negative', 'neutral', 'not applicable']

# 1. Bar charts for positive, negative, and neutral sentiments (including "not applicable" in percentage)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Loop over the aspects to create a bar chart for each aspect
for i, aspect in enumerate(aspects):
    ax = axes[i]
    # Count the distribution of sentiment labels for the current aspect
    aspect_counts = df7[aspect].value_counts()
    print(f"Sentiment distribution for {aspect}:")
    print(aspect_counts)
    
    # Ensure all categories are included, even if some have 0 counts
    aspect_counts = aspect_counts.reindex(labels, fill_value=0)
    
    # Calculate the percentage for positive, negative, neutral, and "not applicable"
    total_count = aspect_counts.sum()
    sentiment_percentage = (aspect_counts / total_count) * 100
    
    # Plot the bar chart for this aspect (including "not applicable" as a percentage)
    ax.bar(sentiment_percentage.index, sentiment_percentage.values, color='skyblue')
    
    # Add percentage text at the bottom of each bar, including 0%
    for idx, value in enumerate(sentiment_percentage.values):
        ax.text(sentiment_percentage.index[idx], value + 2, f'{value:.1f}%', ha='center', color='black')
    
    # Set labels and title
    ax.set_title(f'{aspect.capitalize()} Sentiment Distribution')
    ax.set_xlabel('Sentiment')
    ax.set_ylabel('Percentage (%)')
    ax.set_xticks(list(sentiment_percentage.index))
    ax.set_xticklabels(list(sentiment_percentage.index), rotation=45)

    # Add legend
    ax.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()

# 2. Bar chart for "Not Applicable" sentiment only (in percentage)
fig, ax = plt.subplots(figsize=(8, 6))

# Initialize a list to hold the "Not Applicable" counts for each aspect
na_counts = []

# Loop through the aspects and count the occurrences of "Not Applicable"
for aspect in aspects:
    na_count = df7[aspect].value_counts().get('not applicable', 0)
    na_counts.append(na_count)

# Calculate the percentage for "Not Applicable" sentiment relative to total reviews
na_percentage = [(count / len(df7)) * 100 for count in na_counts]

# Plot the bar chart for "Not Applicable" sentiment (in percentage)
ax.bar(aspects, na_percentage, color='gray')
ax.set_title('Not Applicable Sentiment Distribution (Percentage)')
ax.set_xlabel('Aspect')
ax.set_ylabel('Percentage (%)')
ax.set_xticklabels(aspects, rotation=45)

# Add percentage text at the bottom of each bar, including 0% if applicable
for i, percentage in enumerate(na_percentage):
    ax.text(i, percentage + 0.5, f'{percentage:.1f}%', ha='center', color='black')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot for "Not Applicable" bar chart
plt.show()
```

```python
# Combine all DataFrames into one
combined_data = []
for df, name in dataframes:
    if 'Sentiment' in df.columns:
        sentiment_counts = df['Sentiment'].value_counts()
        for sentiment, count in sentiment_counts.items():
            combined_data.append({'Model': name, 'Sentiment': sentiment, 'Count': count})

# Convert combined data to a DataFrame
combined_df = pd.DataFrame(combined_data)

# Plotting the combined data
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=combined_df,
    x='Model',
    y='Count',
    hue='Sentiment',
    style='Sentiment',
    s=100,  # Size of the dots
    palette='viridis'
)
plt.title('Sentiment Occurrences Across Models')
plt.xlabel('Model')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```
