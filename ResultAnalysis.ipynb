```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('1.distilbert.csv')
df2 = pd.read_csv('2.roberta.csv')
df3 = pd.read_csv('3.robertaXL.csv')
df4 = pd.read_csv('4.McDSentimentAnalysis.csv')
df5 = pd.read_csv('5.multiLingual.csv')
df6 = pd.read_csv('6.AspectBasedSentimentAnalysis.csv')
df7 = pd.read_csv('6.AspectBasedSentimentAnalysis_withApplicability.csv')
```

```python
# List of DataFrames and their names
dataframes = [
    (df1, "DistilBERT"),
    (df2, "RoBERTa"),
    (df3, "RoBERTa XL"),
    (df4, "McD Sentiment Analysis"),
    (df5, "Multilingual"),
    (df6, "Aspect-Based Sentiment Analysis"),
    (df7, "Aspect-Based Sentiment Analysis with Applicability"),
]
```

```python
# Function to analyze and plot sentiment distribution with percentages
def analyze_and_plot_sentiment(df, title):
    # Check if the DataFrame is not empty
    if df.empty:
        print(f"{title} DataFrame is empty. Skipping...")
        return
    
    # Check if the 'Sentiment' column exists
    if 'Sentiment' not in df.columns:
        print(f"'Sentiment' column not found in {title}. Skipping...")
        return
    
    # Display sentiment distribution
    sentiment_counts = df['Sentiment'].value_counts()
    total = sentiment_counts.sum()
    percentages = (sentiment_counts / total) * 100
    print(f"Sentiment distribution for {title}:")
    print(sentiment_counts)
    print(f"Percentages:\n{percentages}")
    
    # Plot sentiment distribution
    plt.figure(figsize=(8, 6))
    ax = sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="viridis")
    plt.title(f"{title} - Sentiment Distribution")
    plt.xlabel("Sentiment")
    plt.ylabel("Count")
    
    # Annotate bars with percentages
    for i, count in enumerate(sentiment_counts.values):
        percentage = f"{percentages.iloc[i]:.1f}%"
        ax.text(i, count + 0.5, percentage, ha='center', va='bottom', fontsize=10, color='black')
    
    plt.show()

# Generate sentiment distribution diagrams for each DataFrame
for df, name in dataframes:
    analyze_and_plot_sentiment(df, name)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
aspects = ['feature', 'ads', 'performance']
colors = ['green', 'yellow', 'red']

for ax, aspect in zip(axes, aspects):
    data = df6[aspect].value_counts()
    ax.pie(
        data,
        labels=data.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors
    )
    ax.set_title(f'Sentiment for {aspect.capitalize()}')

fig.text(0.5, 0.01, 'Sentiment Proportions by Aspect', ha='center', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Combine all DataFrames into one
combined_data = []
for df, name in dataframes:
    if 'Sentiment' in df.columns:
        sentiment_counts = df['Sentiment'].value_counts()
        for sentiment, count in sentiment_counts.items():
            combined_data.append({'Model': name, 'Sentiment': sentiment, 'Count': count})

# Convert combined data to a DataFrame
combined_df = pd.DataFrame(combined_data)

# Plotting the combined data
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=combined_df,
    x='Model',
    y='Count',
    hue='Sentiment',
    style='Sentiment',
    s=100,  # Size of the dots
    palette='viridis'
)
plt.title('Sentiment Occurrences Across Models')
plt.xlabel('Model')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```
